```python
# Import librearies for data manipulation and cleaning
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Load the data
data = pd.read_excel('data.xlsx')
```

```python
# Show the first 5 rows of the data
data.head()
```

```python
# Show the last 5 rows of the data
data.tail()
```

```python
# Show the shape of the data
print("The data has {} rows and {} columns".format(data.shape[0], data.shape[1]))
```

```python
# Show the information of the data
data.info()
```

```python
# Show the summary statistics of the data
data.describe()
```

```python
# Check duplicates
print("the data has {} duplicates".format(data.duplicated().sum()))
```

```python
# Check missing values
print("The data has {} missing values".format(data.isnull().sum()))
```

```python
# Remove columns with many missing values
threshold = 0.5  # Threshold for allowed missing values (50%)
for column in data.columns:
    if data[column].isnull().sum() / len(data) > threshold:
        data.drop(column, axis=1, inplace=True)

# Replace remaining missing values with mean or mode
for column in data.columns:
    if data[column].dtype == 'object':
        # Replace missing values with mode for categorical columns
        data[column].fillna(data[column].mode()[0], inplace=True)
    else:
        # Replace missing values with mean for numerical columns
        data[column].fillna(data[column].mean(), inplace=True)

# Verify the changes and show dropped columns
print("The data has {} rows and {} columns after cleaning".format(data.shape[0], data.shape[1]))
print("The data has {} missing values after cleaning".format(data.isnull().sum().sum()))
print("Columns dropped: {}".format(set(data.columns).symmetric_difference(set(pd.read_excel('data.xlsx').columns))))

# Show the first 5 rows of the cleaned data
data.head()
```

```python
# Boxplots para detectar outliers
print("\nBoxplots para detectar outliers:")

# Configuración de estilo de Seaborn
sns.set(style="whitegrid")

# Generar boxplots para cada columna numérica
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=data[column], palette="Set2")
    plt.title(f'Boxplot de {column}', fontsize=15)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Valores', fontsize=12)
    plt.show()
```

```python
# Correlation matrix includes only numerical columns and without the values

# Set Seaborn style
sns.set_style(style="white")

# Generate correlation matrix
plt.figure(figsize=(12, 6))
sns.heatmap(data.select_dtypes(include=['float64', 'int64']).corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Matrix', fontsize=15)
plt.show()
```

## Correlation Matrix Analysis

The correlation matrix shows the relationship between different variables in the dataset. Here are some key points:

### Strong Positive Correlations:
- **💼 Salario Anual Inicial 2020 and Salario Anual Actual 2020** have a very high correlation (0.995), which is expected since the current salary is likely based on the initial salary.
- **⏰ Horas Jornada and FTE** have a very high correlation (0.977), indicating that these two variables are closely related.
- **📊 Cat Lab num and Salario Anual Inicial 2020** have a high correlation (0.898), suggesting that the job category significantly influences the initial salary.

### Strong Negative Correlations:
- **👶 Nuevas contrataciones 2020 and Edad** have a moderate negative correlation (-0.354), which could indicate that new hires tend to be younger.
- **📉 Nuevas contrataciones 2020 and Antigüedad Meses** have a moderate negative correlation (-0.371), which is logical since new employees have less tenure.

### Moderate Correlations:
- **🎓 Formación Oficial Num and Rotación Interna** have a moderate correlation (0.535), suggesting that official training may be related to internal turnover.
- **📚 Horas de formación recibidas and Salario Anual Inicial 2020** have a moderate correlation (0.787), indicating that more training hours may be associated with a higher initial salary.

### General Observations:
- Most correlations between **🆔 ID Empleado** and other variables are low, which is expected since ID Empleado is an identifier variable.
- **📈 Evaluación Desempeño** does not show strong correlations with any other variable, which could indicate that performance is not strongly influenced by the available variables in the dataset.
